#**Matplotlib Library: module: PyPlot & Seaborn**

```
data=load_iris()
df_iris=pd.DataFrame(data=data.data,columns=data.feature_names)
df_iris.head()
```

```
df_iris_copy=df_iris.copy()
```

```
df_iris.concat(df_iris_copy)
```

```
pd.merge(df_iris, df_iris_copy, on='petal width (cm)')
```

```
#applymap used on dataframe ONLY
df_iris['sepal width (cm)']=df_iris.applymap(lambda x: x*2)
df_iris.head()
```

```
df_c_s_p=df[['Pclass','Sex','PassengerId']].groupby(['Pclass','Sex'],as_index=False).count()
print('Count passengers per PClass')
print(df_c_s_p)
print('---------')
print('Pivoting new dataframe')
df_pivot = df_c_s_p.pivot(columns='Pclass',index='Sex')
print(df_pivot)
print('---------')
df_pivot.plot(kind='bar',stacked=True,legend=True,title='Sex/PClass distribution')
plt.show()
```

```
#calculate percentage 
df_c_s_p['Tot_Sex']=df_c_s_p[['Sex','PassengerId']].groupby(['Sex']).transform('sum')
df_c_s_p['Perc_Sex']=df_c_s_p['PassengerId']/df_c_s_p['Tot_Sex']
print('Adding percentage Sex/Class')
print(df_c_s_p)
print('---------')
df_pivot = df_c_s_p.pivot(columns='Pclass',index='Sex',values='Perc_Sex')
#plot it
print('Percentage Sex/Class')
print(df_pivot)
print('---------')

df_pivot.plot(kind='bar',stacked=True,legend=True,title='Percentage Sex/PClass',figsize=(10,6))
plt.show()
```

```
plt.figure(figsize=(10,6))
plt.plot(df_iris,marker='s')

plt.legend(loc='best')

plt.title("Iris")

plt.show()
```

```
plt.figure(figsize=(10,6))

plt.plot(df_iris['sepal length (cm)'],color="r",label="sepal length (cm)",marker='o')
plt.plot(df_iris['sepal width (cm)'],color='b',label="sepal width (cm)",marker='h')
plt.plot(df_iris['petal length (cm)'],color='g',label="petal length (cm)",marker='s')
plt.plot(df_iris['petal width (cm)'],color='y',label="petal width (cm)",marker='h')

plt.legend(loc='best')

plt.title("Iris")

plt.show()
```

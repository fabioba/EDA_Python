```
#drop column
df.drop(['Test'],
        axis=1,
        inplace=True)
df.head()
```

```
#map function -> used on Series
df['2class']=df['Pclass'].map(lambda x: x*2)
df.head()
```

```
#apply function->used on Dataframe and Series
def double_class(item):
  double_item=item['Pclass']*2
  return double_item

df['2class_apply']=df[['Pclass','Sex']].apply(lambda x:double_class(x),
                                              axis=1)
df.head()
```

```
#https://matplotlib.org/tutorials/introductory/usage.html#sphx-glr-tutorials-introductory-usage-py
fig = plt.figure()  # an empty figure with no axes
fig.suptitle('No axes on this figure')  # Add a title so we know which it is

fig, ax_lst = plt.subplots(2, 2)  # a figure with a 2x2 grid of Axes
```

```
#Charts overview

fig,ax=plt.subplots(2,2,            #ax = boxes which include the plots
                    figsize=(15,10))
fig.suptitle('Quickview Plots')

df['Age'][df['Age'].isnull()==False].plot(kind='line',
                                          ax=ax[0,0],
                                          linestyle='--',
                                          linewidth=3,
                                          color='r',
                                          marker='o')
ax[0,0].set_title("Lineplot")

df['Age'][df['Age'].isnull()==False].plot(kind='hist',
                                                  ax=ax[0,1],color='g')
ax[0,1].set_title("Histogram")

df['Age'][df['Age'].isnull()==False].plot(kind='bar',
                                          ax=ax[1,0],color='b')
ax[1,0].set_title("Barplot")

df['Age'][df['Age'].isnull()==False].plot(kind='area',
                                          ax=ax[1,1])
ax[1,1].set_title("Area")

plt.show()#show image
```

# **LINEPLOT**

```
fig=plt.figure(figsize=(10,6))


plt.title('Fare Plot')
plt.xlabel('Freq')
plt.ylabel('Fare_Value')

df['Fare'].plot(kind='line',
                color='r',
                alpha=0.6,#transparency
                linewidth=2,
                marker='o',
                linestyle='--')
plt.ylim(0,300)
plt.xlim(0,50)#limit values along the x-axis
#plt.xticks([0,20,30,50]) #set all values along the x-axis

plt.show()
```

```
f, axes = plt.subplots(2, 1,          #numrows,numcols subplots
                       figsize=(10,10))
f.suptitle('Distribution')
sns.lineplot(x='Age', 
             y="Fare", 
             hue="Pclass",
              data=df,ax=axes[0]).set_title("Age/Fare")

sns.lineplot(x='Age', y="Fare", hue="Sex",
              data=df,ax=axes[1])
plt.show()
```

# **BARPLOT**

```
#count distinct values
df['Sex'].value_counts()
```

```
figure=plt.figure(figsize=(10,6))
df['Survived'].value_counts().plot(kind='bar',
                color=['r','b'],
                alpha=0.5)
plt.title('Survived Bar')
plt.show()
```

```
fig=plt.figure(figsize=(10,6))
sns.countplot(data=df, 
              x='Pclass',
              hue='Sex', 
              palette='magma')
plt.title('PCLASS/SEX')
plt.show()
```

# **Distribution**

```
fig=plt.figure(figsize=(10,6))
plt.title('Distribution')
sns.distplot(df['Fare'])
plt.show()
```

```
fig=plt.figure(figsize=(10,10))
df_subplot=df[['Fare','Pclass','SibSp','Parch']]
i=1
for item in df_subplot:
  plt.subplot(2,2,i,title=item)
  sns.distplot(df_subplot[item])
  i=i+1
plt.show()
```

# **Area**

```
df['Fare2']=df['Fare'].map(lambda x: x*1.2)

df[['Fare','Fare2']].plot(kind='area',
                          color=['red','yellow',],
                          alpha=0.5,
                          figsize=(10,5))
plt.title('Area')

plt.xlabel('Fare')
plt.show()
```

# **Histogram**

```
plt.figure(figsize=(10,5))
df['Fare'].plot(kind='hist',
                edgecolor='black',
                alpha=0.4,
                bins=10)

plt.show()
```

# **Boxplot**

```
fig=plt.figure(figsize=(10,5))
plt.subplot(1,2,1)#row,col,position
df['Fare'].plot(kind='box',
                color='blue')


plt.subplot(1,2,2)#row,col,position
df['Age'][df['Age'].isnull()==False].plot(kind='box',
                                          color='green')

plt.show()
```

```
plt.figure(figsize=(10,5))
sns.boxplot(x='Survived',
            y='Age',
            hue='Pclass',
            data=df[df['Age'].isnull()==False])
plt.show()
```

# **Heatmap Correlation**

```
#Pearson correlation
fig=plt.figure(figsize=(10,6))
fig.suptitle('Pearson Correlation')
sns.heatmap(df[['Age','Pclass','Survived']].corr())
plt.show()
```

# **Regression Plot**

```
plt.figure(figsize=(10,6))
plt.title('Regression')
sns.regplot(x='Fare',y='Age',data=df[df['Age'].isnull()==False],marker='+',color='red')
plt.show()
```

#**Matplotlib Library: module: PyPlot & Seaborn**

```
#https://matplotlib.org/tutorials/introductory/usage.html#sphx-glr-tutorials-introductory-usage-py
fig = plt.figure()  # an empty figure with no axes
fig.suptitle('No axes on this figure')  # Add a title so we know which it is

fig, ax_lst = plt.subplots(2, 2)  # a figure with a 2x2 grid of Axes
```

```
fig=plt.figure(figsize=(10,10))

fig.suptitle('Quickview Plots')
fig,ax=plt.subplots(2,2,sharex=True,figsize=(15,10))#ax corresponds to the box which includes the plot

df['Age'][df['Age'].isnull()==False].plot(kind='line',
                                          ax=ax[0,0],linestyle='--',
                                          linewidth=3,
                                          color='r',
                                          marker='o')
ax[0,0].set_title("Lineplot")

df['Age'][df['Age'].isnull()==False].plot(kind='hist',
                                                  ax=ax[0,1],color='g')
ax[0,1].set_title("Histogram")

df['Age'][df['Age'].isnull()==False].plot(kind='bar',
                                          ax=ax[1,0],color='b')
ax[1,0].set_title("Barplot")

df['Age'][df['Age'].isnull()==False].plot(kind='area',
                                          ax=ax[1,1])
ax[1,1].set_title("Area")

plt.show()
```

# **LINEPLOT**

```
fig=plt.figure(figsize=(10,6))


plt.title('Fare Plot')
plt.xlabel('Freq')
plt.ylabel('Fare_Value')

df['Fare'].plot(kind='line',
                color='r',
                alpha=0.6,
                linewidth=2,
                marker='o',
                linestyle='--')
plt.ylim(0,300)
plt.xlim(0,50)

plt.show()
```

```
f, axes = plt.subplots(2, 1,figsize=(10,10))#numrows,numcols
f.suptitle('Distribution')
sns.lineplot(x='Age', y="Fare", hue="Pclass",
data=df,ax=axes[0])

sns.lineplot(x='Age', y="Fare", hue="Sex",
data=df,ax=axes[1])
plt.show()
```

# **BARPLOT**

# **Piechart**

```
df_test_sum=df[['PassengerId','Pclass','Sex']].groupby(['Pclass','Sex']).count().reset_index()
print(df_test_sum)
label_list=df_test_sum[['Pclass','Sex']].values
print(label_list)
label_list_str=[str(x) for x in label_list]

df_test_sum.plot.pie(y=['PassengerId'],
                     #colors=['orange','lightblue','lightgreen','red','gold','green'],
                     shadow=True,
                     startangle=90,#explode=(0.1,0,0,0,0,0),#explode first slice
                     wedgeprops={"edgecolor":"k",'linewidth': 1, 'linestyle': '-', 'antialiased': True},
                     subplots=True,
                     figsize=(12,8)
                 )
plt.title('Sex')
plt.legend(label_list_str)
plt.show()
```

# **Heatmap Correlation**

```
#Pearson correlation
fig=plt.figure(figsize=(10,6))
fig.suptitle('Pearson Correlation')
sns.heatmap(df[['Age','Pclass','Survived']].corr())
plt.show()
```

# **Regression Plot**

```
plt.figure(figsize=(10,6))
plt.title('Regression')
sns.regplot(x='Fare',y='Age',data=df[df['Age'].isnull()==False],marker='+',color='red')
plt.show()
```
